```python
import cv2 as cv
import numpy as np
import pandas as pd
print(cv.__version__)
```

### Load Data

```python
df = pd.read_csv('data/train.csv')
```

### Explore Data

```python
df.describe()
```

### Preprocess Data

```python
# split train and test data
split_rate = 0.8
split_sep = int(split_rate * df.shape[0])
train_df = df.iloc[: split_sep, :]
test_df = df.iloc[split_sep:, :]
```

```python
def preprocess_labels(df):
    nrows = df.shape[0]
    return df.to_numpy().reshape(nrows, 1)

def preprocess_data(df):
    nrows = df.shape[0]
    ncols = df.shape[1]
    return df.to_numpy().reshape(nrows, ncols).astype(np.float32) / 255.0
```

```python
train_labels = preprocess_labels(train_df['label'])
train_data = preprocess_data(train_df.drop('label', axis=1))
```

```python
print(type(train_labels))
print(train_labels.shape)

print(type(train_data))
print(train_data.shape)
```

```python
test_labels = preprocess_labels(test_df['label'])
test_data = preprocess_data(test_df.drop('label', axis=1))
```

```python
knn = cv.ml.KNearest_create()
knn.train(train_data, cv.ml.ROW_SAMPLE, train_labels)
```

```python
# Accuracy of train data
ret,result,neighbours,dist = knn.findNearest(train_data,k=5)
matches = result==train_labels
correct = np.count_nonzero(matches)
accuracy = correct*100.0/result.size
print(accuracy)
```

```python
# Accuracy of test data
ret,result,neighbours,dist = knn.findNearest(test_data,k=5)
matches = result==test_labels
correct = np.count_nonzero(matches)
accuracy = correct*100.0/result.size
print(accuracy)
```
